```python
# Only keep the following columns: 'Intervention', 'PedPed', 'Barrier', 'CrossingSignal','NumberOfCharacters', 'DiffNumberOFCharacters','Saved','Man', 'Woman', 'Pregnant', 'Stroller','OldMan', 'OldWoman', 'Boy', 'Girl', 'Homeless', 'LargeWoman','LargeMan', 'Criminal', 'MaleExecutive', 'FemaleExecutive', 'FemaleAthlete', 'MaleAthlete', 'FemaleDoctor', 'MaleDoctor', 'Dog', 'Cat'
data = data[['ResponseID','Intervention', 'PedPed', 'Barrier', 'CrossingSignal','NumberOfCharacters', 'DiffNumberOFCharacters','Saved','Man', 'Woman', 'Pregnant', 'Stroller','OldMan', 'OldWoman', 'Boy', 'Girl', 'Homeless', 'LargeWoman','LargeMan', 'Criminal', 'MaleExecutive', 'FemaleExecutive', 'FemaleAthlete', 'MaleAthlete', 'FemaleDoctor', 'MaleDoctor', 'Dog', 'Cat']]
data.head()
```

```python
# sort the data by ResponseID
data = data.sort_values('ResponseID')
data.head(6)
```

```python
comparisons = []

# get unique response ids
response_ids = data['ResponseID'].unique()

# each response has 2 rows, one for the chosen and one for the unchosen. The chosen row is the one with saved = 1
# We need to make a list of lists called comparisons where the inner list has the following format: [chosen_row_values_other_than_response_id_and_saved, unchosen_row_values_other_than_response_id_and_saved]

for response_id in tqdm(response_ids):
    response_data = data[data['ResponseID'] == response_id]
    chosen_row = response_data[response_data['Saved'] == 1]
    unchosen_row = response_data[response_data['Saved'] == 0]

    columns_to_keep = [col for col in data.columns if col not in ['ResponseID', 'Saved']]
    # print(list(chosen_row[columns_to_keep].values[0]))
    # print(list(unchosen_row[columns_to_keep].values[0]))

    comparisons.append([list(chosen_row[columns_to_keep].values[0]), list(unchosen_row[columns_to_keep].values[0])])
print(comparisons[0])
    
```

```python
# save this in a csv file
comparisons_df = pd.DataFrame(comparisons, columns=['Chosen', 'Unchosen'])
comparisons_df.to_csv('comparisons.csv', index=False)
```
