```python
import pandas as pd
import numpy as np
```

```python
# set max columns and rows to display
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

```python
mm_df = pd.read_csv('Moral Machine Data/SharedResponses.csv')
```

```python
mm_df.head()
```

```python
response_ids = mm_df['ResponseID'].unique()
```

```python
print('Number of unique responses:', len(response_ids))
```

```python
# sample 100,000 response ids
shuffled_response_ids = np.random.permutation(response_ids)
sampled_response_ids = shuffled_response_ids[:100000]
```

```python
# filter the dataframe to only include the sampled response ids
sampled_mm_df = mm_df[mm_df['ResponseID'].isin(sampled_response_ids)]
```

```python
sampled_mm_df = sampled_mm_df.groupby('ResponseID').filter(lambda x: len(x) > 1)
```

```python
# save this sampled dataframe to a csv file
sampled_mm_df.to_csv('Moral Machine Data/SharedResponsesSample.csv', index=False)
```
